```python
#station with no price information 
#price information with no station infor

station_id=df_station.ID.unique()
price_stid=df_price.STID.unique()

removeFromStationInfo=[]
for i in station_id:
    if i not in price_stid:
        print(i)
        removeFromStationInfo.append(i)
print("--------------------")

removeFromPriceInfo=[]
for j in price_stid:
    if j not in station_id:
        print(j)
        removeFromPriceInfo.append(j)

        
           
```

```python
for priceId in removeFromPriceInfo:
    df_price.drop(df_price[df_price['STID']==priceId].index,axis=0,inplace=True)

for s_id in removeFromStationInfo:
    df_station.drop(df_station[df_station['ID']==s_id].index,axis=0,inplace=True)

```
